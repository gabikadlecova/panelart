```python
def process_row(r, with_nuclear=False):
  if 'ja' not in r['voted'] and 'nein' not in r['voted']:
    return None

  if 'nicht in' in r['party_voted']:
    return None

  res = {}
  res['gender'] = r['gender'] if r['gender'] == 'weiblich' else 'männlich'
  res['age'] = 2017 - r['birthyear'] if isinstance(r['birthyear'], int) else None
  res['voted'] = 'ja' in r['voted']
  res['ostwest'] = r['ostwest']

  if r['bc'] == 'gennant' or r['mgr'] == 'gennant':
    res['education'] = 'einen Hochschulabschluss'
  else:
    for k, v in edu_map.items():
      if k in r['education']:
        res['education'] = v
        break

  if 'bis' in r['income']:
    income = r['income'].replace(' Euro', '').split(' bis unter ')
    upper = int(income[1])
    if upper < income_lims[0]:
      res['income'] = 'niedriges'
    elif upper < income_lims[1]:
      res['income'] = 'mittleres'
    else:
      res['income'] = 'hohes'
  else:
    res['income'] = None

  vot = r['party_voted']
  if 'keine' in vot or 'ungue' in vot or 'weiss' in vot or 'trifft' in vot:
    res['voted'] = None
  else:
    res['voted'] = vot

  if res['voted'] is None and 'nein' in r['voted']:
    res['voted'] = 'no vote'

  def to_num(k, val, m):
    if str.isdigit(val[0]):
      if len(val) > 1:
        val = val[:2]
      res[k] = m[int(val)]
    else:
      res[k] = None

  to_num('linksrechts', r['linksrechts'], linksmap)
  to_num('angst_migrants', r['angst_migrants'], angstmap)
  if with_nuclear:
    to_num('angst_nuclear', r['nuclear'], angstmap)

  for k, v in limmig_map.items():
    if k in r['limit_migrants']:
      res['limit_migrants'] = v
      break

  for k, v in ineq_map.items():
    if k in r['commies']:
      res['reduce_ineq'] = v
      break

  if 'Ausbild' in r['job']:
    res['job'] = 'in Ausbildung'
  elif 'trifft' in r['job']:
    res['job'] = 'nicht berufstätig'
  elif 'keine An' in r['job'] or 'weiss n' in r['job']:
    res['job'] = None
  else:
    res['job'] = 'berufstätig'

  if 'weiss' in r['religious'] or 'keine' in r['religious']:
    res['religious'] = None
  else:
    res['religious'] = r['religious'].replace('oes', 'ös').replace('ueb', 'üb')

  return res
```

```python
def gen_prompt(row):
  ichhabe = {k: row[k] for k in ['education', 'income', 'job'] if row[k] is not None}

  num = 0
  if len(ichhabe) > 0:
    ichtext = 'Ich habe '
    if 'education' in ichhabe:
      ichtext += ichhabe['education']
      num += 1

    if 'income' in ichhabe:
      if len(ichhabe) == 3:
        ichtext += ', '
      elif len(ichhabe) == 2 and num > 0:
        ichtext += " und habe "
      ichtext += f"ein {ichhabe['income']} monatliches Haushalts-Nettoeinkommen"

    if 'job' in ichhabe:
      if len(ichhabe) > 1:
        ichtext += " und bin "
      ichtext += ichhabe['job']

    ichtext += '. '

  age = '' if row['age'] is None else f"Ich bin {row['age']} Jahre alt"
  if len(age) == 0 and row['gender'] is not None:
    agegen = f"Ich bin {row['gender']}. "
  else:
    agegen = age + f" und {row['gender']}" if row['gender'] is not None else age
    agegen += '. '

  religious = '' if row['religious'] is None else f"Ich bin {row['religious']}. "
  links = '' if row['linksrechts'] is None else f"Politisch-ideologisch ordne ich mich {row['linksrechts']} ein. "

  migrants = 'Ich finde, die Regierung sollte'
  initlen = len(migrants)
  if row['limit_migrants'] is not None:
    migrants += f" die Einwanderung {row['limit_migrants']}"

  if row['reduce_ineq'] is not None:
     if len(migrants) > 0:
      migrants += " und "
     migrants += f"{row['reduce_ineq']}, um die Einkommensunterschiede zu verringern"

  if len(migrants) > initlen:
    migrants += '.'

  angst = '' if row['angst_migrants'] is None else f" Ich habe {row['angst_migrants']} von Flüchtlingen."
  if 'angst_nuclear' in row:
    angst += '' if row['angst_nuclear'] is None else f" Ich habe {row['angst_nuclear']} von Kernenergie."

  prompt = f"""
  In place of [INSERT], fill in (in German) whether the respondent voted and if yes, then for what party. If unsure, list the probable parties with probabilities (always output whether the respondent voted and for what party).
  List as many parties as necessary. Optionally answer \"andere Partei\" if the voter voted for a small unpopular party.

  The output format is:
    [gewählt, proba a], [nicht gewählt, proba b]; [PARTY1, proba 1], [PARTY2, proba2],...
  "
  {agegen}{ichtext}{religious}{links}{migrants}{angst} Ich lebe in {row['ostwest']}. Habe ich bei der
  Bundestagswahl 2017 gewahlt und wenn ja, welcher Partei
  habe ich meine Zweitstimme gegeben? Ich habe [INSERT]"
  """
  return prompt
```

```python
def parse_response(t):
  voted, parties = t.split('; ')
  parties = parties[1:-1].split('], [')

  voted = voted.replace('[', '').replace(']', '').split(', ')
  if len(voted) == 2:
    pv = float(voted[1])
    if voted[0] == 'gewählt':
      voted = [pv, 1.0 - pv]
    elif voted[0] == 'nicht gewählt':
      voted = [1.0 - pv, pv]
    else:
      raise ValueError(f"Invalid voted: {voted}")

  else:
    assert voted[0] == 'gewählt' and voted[2] == "nicht gewählt"
    voted = [float(voted[1]), float(voted[3])]

  pnames = []
  pbs = []
  for p in parties:
    #print(p)
    name, pb = p.split(', ')
    pb = pb.replace('proba ', '')
    pnames.append(name)
    pbs.append(float(pb))
  return pnames, pbs, voted

responses['988']
```

```python
def to_valid_n(v):
  if 'Grün' in v or 'GRÜN' in v or 'GRU' in v:
    v = "Grüne"
  if 'CDU' in v or 'CSU' in v:
    v = 'CDU/CSU'
  if 'SPD' in v:
    v = 'SPD'
  if 'LINKE' in v or 'Linke' in v:
    v = 'Die Linke'
  if v == 'AFD' or v == 'AfD':
    v = 'AfD'
  if 'ANDERE' in v or 'Andere' in v:
    v = 'andere Partei'
  return v
```

```python
import numpy as np
from collections import Counter

pred_vote = {}
did_vote = []

n_samples = 10
res_dfs = [{} for _ in range(n_samples)]

for i, v in responses.items():
  i = int(i)
  text = v.text if hasattr(v, 'text') else v

  if '[' not in text:
    print('The LLM is totally dumb.', flush=True)  # frustrated output after sanitizing outputs...
    continue

  if text.startswith('gew'):
    text = f"[{text}"
  text = text.replace('\"', '')

  if text.endswith(';') or text.endswith('.') or text.endswith(']]'):
    text = text[:-1]

  text = text.replace('0,', '0.')
  text = text.replace('<0.1', '0.0')

  text = text.split('\n')[0]
  text = text.replace(', []', '')
  if 'nicht' not in text:
    text = text.replace("gewählt, 1.0],", "gewählt, 1.0];")

  pnames, pbs, voted = parse_response(text)
  pbsum = sum(pbs)
  if pbsum != 1:
    pbs = [p / pbsum for p in pbs]

  for idx in range(n_samples):
    did_vote.append(voted)
    party = np.random.choice(pnames, size=1, p=pbs)

    vote_choice = np.random.choice([True, False], size=1, p=voted)
    pred_vote[i] = str(party[0]) if vote_choice else 'no vote'

    def save_to_res(res_df, key, val):
        if val not in res_df:
            res_df[val] = {}
        if key not in res_df[val]:
            res_df[val][key] = 0
        res_df[val][key] += 1


    the_vote = new_data.loc[int(i)]['voted']
    if the_vote is None:
      continue

    if 'Die Par' in str(party[0]) or 'Die Par' in the_vote:
      continue

    the_vote = to_valid_n(the_vote) if the_vote is not None else 'no vote'
    save_to_res(res_dfs[idx], 'voted', the_vote)
    save_to_res(res_dfs[idx], 'predicted', to_valid_n(pred_vote[i]))
```
